```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.support.select import Select

import time
import pandas as pd
import numpy as np
```

```python
def page_load_id(element, delay):
    '''Delays the webdriver  by the variable (delay) until (element) is loaded'''
    try:
        myElem = WebDriverWait(
            driver, delay
            ).until(
            expected_conditions.presence_of_element_located(
                (By.ID, element)
            ))
        print("Page is ready!")
    except TimeoutException:
        print("Loading took too much time!")

def page_load_class(element, delay):
    '''Delays the webdriver  by the variable (delay) until (element) is loaded'''
    try:
        myElem = WebDriverWait(
            driver, delay
            ).until(
            expected_conditions.presence_of_element_located(
                (By.CLASS_NAME, element)
            ))
        print("Page is ready!")
    except TimeoutException:
        print("Loading took too much time!")
```

```python
driver = webdriver.Chrome('c:/Users/Ming/downloads/chromedriver')
driver.get('https://top100.winespectator.com/lists/')
time.sleep(5)
```

```python
wine_df = pd.DataFrame(columns=['wine','wine_notes', 'vintage'])

for year in range(22):
    page = year + 1
    print(f'Start page {page}:')
    select = Select(driver.find_element_by_xpath("/html/body/main/div[5]/div[1]/select")) # Select year dropdown
    try:
        select.select_by_index(year)
    except NoSuchElementException:
        print('The year is not available.')

    driver.refresh()
    driver.find_element_by_xpath('//div[@class="whatever"]').click() # Clicks on expand all button
    
    wine = driver.find_elements_by_xpath('//span[@class="wineName"]')
    wines_notes = driver.find_elements_by_xpath('//div[@class="tabel-note"]')
    vintages = driver.find_elements_by_xpath('//td[@class="vintage"]')
    
    wine_list = []
    for p in range(len(wine)):
        wine_list.append(wine[p].text)
    print(f'wine_list scrapped: {page}')

    wine_note_list = []
    for p in range(len(wines_notes)):
        wine_note_list.append(wines_notes[p].text)
    print(f'wine_note scrapped: {page}')
        
    vintage_list = []
    for p in range(len(vintages)):
        vintage_list.append(vintages[p].text)
    print(f'vintages scrapped: {page}')
    
    data_tuple = list(zip(wine_list, wine_note_list, vintage_list))
    temp_df = pd.DataFrame(data_tuple, columns=['wine','wine_notes', 'vintage'])
    wine_df = pd.concat(objs = [wine_df, temp_df])
    print(f'wine_df complete: {page}')
    
    print(f'Page {page} done.')
    print(f'{len(wine)} wines scrapped.')
    print('----------')
    if page == 1:
        print('DataFrame Output Example')
        print(temp_df)
        print('----------')
```

```python
def scrape_vinuous(start, stop, df):
    new_df = df
    for page in range(start,stop):

        driver.get(f'https://www.vinous.com/wines?advanced=true&page={page}')

        page_load_id('new_wine_list', 10) # Wait for Carousel item to load

        wine = driver.find_elements_by_css_selector('td[data-title="Name"]>a')
        wine_notes = driver.find_elements_by_css_selector('td[data-title="Tasting Note"]>p')
        vintages = driver.find_elements_by_css_selector('td[data-title="Vintage"]')

        if len(wine) != 0:  # Check to see if table has info/404 check
            wine_list = []
            for p in range(len(wine)):
                wine_list.append(wine[p].text)
            print(f'wine_list scrapped: page {page}')

            wine_note_list = []
            for p in range(len(wine_notes)):
                wine_note_list.append(wine_notes[p].text)
            print(f'wine_note scrapped: page {page}')

            vintage_list = []
            for p in range(len(vintages)):
                vintage_list.append(vintages[p].text)
            print(f'vintages scrapped: page {page}')

            data_tuple = list(zip(wine_list, wine_note_list, vintage_list))
            temp_df = pd.DataFrame(data_tuple, columns=['wine','wine_notes', 'vintage'])
            new_df = pd.concat(objs = [new_df, temp_df])
            print(f'wine_df complete: page {page}')

            print(f'page {page} done.')
            print(f'{len(wine)} wines scrapped.')
            print('----------')
            
        if page == 1:
            print('DataFrame Output Example')
            print(temp_df)
            print('----------')
            
    return new_df
```
